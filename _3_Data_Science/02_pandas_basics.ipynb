```python
# merge() or pd.merge() function is used to combine two dataframes based on the common column
left = pd.DataFrame({"key": ["K0", "K1", "K2", "K3"], "lval": [1, 2, 3, 4]})
right = pd.DataFrame({"key": ["K0", "K1", "K4", "K5"], "rval": [5, 6, 7, 8]})
display(left)
display(right)
```

```python
pd.merge(left, right, on="key", how="inner")   # inner join (default) - only keep rows that have the same key in both dataframes
```

```python
pd.merge(left, right, on="key", how="left")    # left join - keep all rows in the left dataframe
```

```python
pd.merge(left, right, on="key", how="outer")   # outer join - keep all rows in both dataframes
```

```python
left.merge(right, on="key", how="outer")       # right join - keep all rows in the right dataframe
```

```python
# Groupby() function is used to group data based on the values of a column
df.groupby("prefecture").count()
```

```python
df.groupby("prefecture").agg({"store_cd": "count", "floor_area": "mean"})
```

```python
df.groupby("prefecture").agg(
    floor_area_mean=pd.NamedAgg(column="floor_area", aggfunc="mean"),
    floor_area_sum=pd.NamedAgg(column="floor_area", aggfunc="sum"),
    store_cd_count=pd.NamedAgg(column="store_cd", aggfunc="count"),
)
```

```python
# Pivot_table() function is used to create a pivot table
import numpy as np
df1 = pd.DataFrame({
    "index" : ["K0", "K1", "K2", "K3"]*3,
    "column" : ["Tokyo", "Chiba"]*6,
    "C" : np.random.randn(12),
    "D" : np.arange(12)
})

pd.pivot_table(df1, values="C", index=["index"], columns=["column"]).mean(axis=1)
```
